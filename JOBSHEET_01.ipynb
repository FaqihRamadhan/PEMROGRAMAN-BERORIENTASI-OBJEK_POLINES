```python
# Tipe Data Numerik
angka_int = 10
angka_float = 3.14
angka_complex = 2 + 3j

print("Integer:", angka_int, type(angka_int))
print("Float:", angka_float, type(angka_float))
print("Complex:", angka_complex, type(angka_complex))

# Boolean & String
is_active = True
teks = "Hello, Python!"

print("Boolean:", is_active, type(is_active))
print("String:", teks, type(teks))

# Tipe Data Koleksi
daftar_angka = [1, 2, 3, 4, 5]            # List (Mutable)
koordinat = (10, 20)                      # Tuple (Immutable)
himpunan_angka = {1, 2, 3, 2, 1}          # Set (Unique)
data_mahasiswa = {                        # Dictionary (Key-Value)
    "nama": "Andi",
    "nim": "A11.2022.12345",
    "jurusan": "Teknik Informatika"
}

print("List:", daftar_angka)
print("Tuple:", koordinat)
print("Set:", himpunan_angka)
print("Dictionary:", data_mahasiswa)

# Konversi Tipe Data
nilai_str = "100"
nilai_int = int(nilai_str)
print("Konversi String ke Int:", nilai_int, type(nilai_int))
```

```python
# -*- coding: utf-8 -*-

# 1. Pendeklarasian Variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmetika
a = 10
b = 3

penjumlahan = a + b    # Tambah
pengurangan = a - b     # Kurang
perkalian = a * b       # Kali
pembagian = a / b       # Bagi (hasil float)
pembagian_bulat = a // b # Bagi (bulat)
modulus = a % b         # Sisa bagi
pangkat = a ** b        # Pemangkatan

print("\nOPERASI ARITMETIKA")
print("a =", a, ", b =", b)
print("Penjumlahan =", penjumlahan)
print("Pengurangan =", pengurangan)
print("Perkalian =", perkalian)
print("Pembagian =", pembagian)
print("Pembagian Bulat =", pembagian_bulat)
print("Modulus =", modulus)
print("Pangkat =", pangkat)

# 3. Operasi Perbandingan
lebih_besar = a > b
kurang_dari = a < b
sama_dengan = a == b
tidak_sama = a != b
lebih_besar_sama = a >= b
kurang_sama = a <= b

print("\nOPERASI PERBANDINGAN")
print(f"a > b = {lebih_besar}")
print(f"a < b = {kurang_dari}")
print(f"a == b = {sama_dengan}")
print(f"a != b = {tidak_sama}")
print(f"a >= b = {lebih_besar_sama}")
print(f"a <= b = {kurang_sama}")

# 4. Operasi Logika
x = True
y = False

logika_and = x and y
logika_or = x or y
logika_not_x = not x

print("\nOPERASI LOGIKA")
print("x =", x, ", y =", y)
print("x and y =", logika_and)
print("x or y =", logika_or)
print("not x =", logika_not_x)

# 5. Contoh penggunaan di dalam percabangan
if a > b and b > 0:
    print("\nKondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
else:
    print("\nKondisi tidak terpenuhi atau b <= 0.")
```

```python
# -*- coding: utf-8 -*-

# 1. IF Sederhana
nilai = 85
print("Contoh IF Sederhana:")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")

# 2. IF-ELSE
umur = 17
print("Contoh IF-ELSE:")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

# 3. IF-ELIF-ELSE
hari = "Rabu"
print("Contoh IF-ELIF-ELSE:")
if hari == "Senin":
    print("Hari Senin: Saatnya kembali bekerja!")
elif hari == "Selasa":
    print("Hari Selasa: Jadwal rapat mingguan.")
elif hari == "Rabu":
    print("Hari Rabu: Ada diskon di beberapa toko.")
else:
    print("Hari lainnya: Atur jadwalmu dengan baik.\n")

# 4. IF Bersarang (Nested IF)
suhu = 35
print("Contoh IF Bersarang (Nested IF):")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.")
else:
    print("Cuaca sepertinya cukup sejuk.\n")

# 5. Menggabungkan Percabangan dengan Operasi Logika
nilai_teori = 75
nilai_praktik = 80
print("Contoh IF dengan Operasi Logika AND/OR:")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")

# 6. Penggunaan If Ternary
angka = -5
print("Contoh If Ternary (Conditional Expression):")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print("Angka", angka, "=>", status)
```

```python
# -*- coding: utf-8 -*-

# 1. FOR Loop dengan range()
print("1) FOR loop dengan range()")
for i in range(5):
    print("Perulangan ke-", i)
print()

# 2. FOR Loop untuk mengiterasi List
print("2) FOR loop mengiterasi list")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print("Buah:", item)
print()

# 3. WHILE Loop sederhana
print("3) WHILE loop sederhana")
count = 0
while count < 5:
    print("count =", count)
    count += 1
print()

# 4. BREAK pada Loop
print("4) BREAK di dalam loop")
for i in range(10):
    if i == 3:
        print("Loop dihentikan pada i =", i)
        break
    print("i =", i)
print()

# 5. CONTINUE pada Loop
print("5) CONTINUE di dalam loop")
for i in range(5):
    if i == 2:
        print("Lewati i =", i, "dengan continue")
        continue
    print("i =", i)
print()

# 6. NESTED Loop (Loop Bersarang)
print("6) NESTED loop")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
print()

# 7. ELSE pada Loop
print("7) ELSE pada loop for/while")
for x in range(3):
    print("x =", x)
else:
    print("Loop for telah selesai tanpa break.\n")

y = 0
while y < 3:
    print("y =", y)
    y += 1
else:
    print("Loop while telah selesai tanpa break.\n")

# 8. PASS sebagai placeholder
print("8) PASS (placeholder)")
for i in range(3):
    if i == 1:
        pass  # Placeholder, tidak melakukan apa-apa
    print("i =", i)
```

```python
# Program 2: Mengecek Bilangan Genap, Ganjil, dan Prima

# Mengambil input dari pengguna
angka = int(input("Masukkan sebuah bilangan bulat positif: "))

# 1. Mengecek Genap atau Ganjil
# Menggunakan operator modulus (%) untuk sisa bagi [cite: 73, 183]
if angka % 2 == 0:
    status_genap_ganjil = "Genap"
else:
    status_genap_ganjil = "Ganjil"

# 2. Mengecek Bilangan Prima
# Bilangan prima adalah bilangan > 1 yang hanya habis dibagi 1 dan dirinya sendiri
is_prima = True

if angka <= 1:
    is_prima = False
else:
    # Melakukan perulangan untuk mengecek pembagi [cite: 84, 348]
    for i in range(2, int(angka**0.5) + 1):
        if angka % i == 0:
            is_prima = False
            break # Menghentikan loop jika ditemukan pembagi 

# Menentukan keterangan prima
status_prima = "dan merupakan bilangan Prima" if is_prima else "dan bukan merupakan bilangan Prima"

# Output Hasil
print(f"Angka {angka} adalah bilangan {status_genap_ganjil} {status_prima}.")
```
